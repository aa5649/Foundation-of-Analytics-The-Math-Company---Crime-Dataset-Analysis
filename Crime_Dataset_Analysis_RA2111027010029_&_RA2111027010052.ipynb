# Data pre-processing

## finding null values percentage

```python
df.isnull().sum()/df.shape[0] * 100
```

Removing columns with atleast 65% of null values

```python
df = df.drop(['Cross Street','Crm Cd 4','Crm Cd 3','Crm Cd 2','Weapon Desc','Weapon Used Cd'],axis="columns")
```

```python
df
```

```python
df.isnull().sum()
```

# Null values Imputation

```python
df['Vict Sex'].fillna(df['Vict Sex'].mode()[0],inplace=True)
df['Vict Sex'].isna().sum()
```

```python
df['Mocodes'].fillna(df['Mocodes'].mode()[0],inplace=True)
df['Mocodes'].isna().sum()
```

```python
df['Vict Descent'].fillna(df['Vict Descent'].mode()[0],inplace=True)
df['Vict Descent'].isna().sum()
```

```python
df.isnull().sum() / df.shape[0] * 100
```

```python
#discard records with null values in other columns
df = df.dropna()
```

```python
df.isnull().sum()      #Null values have been removed
```

```python
#dropping duplicates
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

## Renaming columns for easier understanding

```python
df.rename(columns = {'DR_NO': 'Report_ID'}, inplace=True)
df.head()
```

```python
df.rename(columns = {'Date Rptd':'Date Reported','DATE OCC':'Date occured','TIME OCC':'Time Occured','AREA':'Area Code','AREA NAME':'Area Name','Rpt Dist No':'District No.','Crm Cd':'Crime code','Crm Cd Desc':
                     'Crime Code Desc'},inplace=True)

df.head()
```

```python
df.rename(columns = {'LON':'Longitude','LAT':'Latitude','LOCATION':'Location','Crm Cd 1':'Additional Crime Code','Mocodes':'Modus Operandi Codes'},inplace=True)
df.head()
```

```python
df['Date Reported']
```

Date Reported column also consists of time as well. It has to be removed.

```python
def remove_time(h):
  h = h[0:10]
  return h

df['Date Reported'] = df['Date Reported'].apply(remove_time)

df.head()
```

```python
#Same with Date occurred
df['Date occured'] = df['Date occured'].apply(remove_time)
df.head()
```

```python
#save to csv
df.to_csv('Crime Dataset.csv',index=False)
```
